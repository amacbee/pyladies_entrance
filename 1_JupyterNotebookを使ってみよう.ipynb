# データ分析で利用されるJupyter Notebook

Jupyter Notebookはデータ分析でよく利用されるものだと紹介しました！  
具体例を紹介します

## 表形式でデータを表現

Excelのように、データを表形式で表現することができます

```python
import numpy as np
import pandas as pd
import scipy as sp

df = pd.DataFrame(np.arange(1, 50, 0.01), columns=['x'])
df['sin_x'] = sp.sin(df.x)

df.head()
```

## データをグラフで描画

Jupyter Notebook上でグラフを確認できる

```python
%matplotlib inline

import matplotlib.pyplot as plt
import scipy as sp
plt.style.use('ggplot')

plt.plot(df.x, df.sin_x)

plt.xlabel('x')
plt.ylabel('sin(x)')
```
